```python
state_code_list = []
site_number_list = []
date_local_list = []
arithmetic_mean_list = []

root_url = 'https://aqs.epa.gov/data/api/dailyData/byState?email=yangwenhui1999@163.com&key=indigomallard75&param=88101&bdate=20200101&edate=20201231&state='
for i in range(1,57):
    print(i)
    if i < 10:
        url = root_url + '0' + str(i)
    else:
        url = root_url + str(i)
    print(url)
    strhtml = requests.get(url)
    soup=BeautifulSoup(strhtml.text,'lxml')
    data = soup.get_text()
    
    for i in re.findall(r'\n.*"state_code": "(.*)",.*\n.*\n.*"site_number": "(.*)",.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*"PM25 24-hour 2012".*\n.*"date_local": "(.*)",.*\n.*\n.*\n.*\n.*\n.*\n.*"arithmetic_mean": (.*),\n', data):
        state_code_list.append(i[0])
        site_number_list.append(i[1])
        date_local_list.append(i[2])
        arithmetic_mean_list.append(i[3])
```

```python
df_dict = {'date':date_local_list, 'FIPS':state_code_list, 'site_number':site_number_list, 'arithmetic_mean':arithmetic_mean_list}
dataframe = pd.DataFrame(data=df_dict)
```

```python
dataframe.to_csv('PM2.5_daily_per_state_20.csv')
```
